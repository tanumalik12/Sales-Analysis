```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("test.csv")

# Bar chart with day against Age
plt.bar(data['Age'], data['Fare'])

plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('Fare')

# Adding the legends
plt.show()
```
